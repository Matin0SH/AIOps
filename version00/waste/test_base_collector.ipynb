```python
from collectors.router_collector import RouterCollector

creds = {
    "username": "matin",
    "password": "cisco",
    "enable_secret": "cisco",
    "timeout": 60,
    "delay_factor": 2,
}

collector = RouterCollector("EDGE-R1", "192.168.56.101", 5000, creds)
collector.connect()

print(collector.get_prompt())

output = collector.send_show_command("show ip interface brief")
print(output)

collector.disconnect()
```

```python
collector.connection_params()
```

```python
# already connected above
print("Enable mode?", collector.check_enable_mode())
print("Config mode?", collector.check_config_mode())

collector.enter_enable_mode()
collector.enter_config_mode()
collector.exit_config_mode()

# single config command
cfg_output = collector.send_config_command("hostname EDGE-R1")
print(cfg_output)

# multiple config commands
cfg_set_output = collector.send_config_set([
    "interface loopback99",
    "description test-loopback",
    "ip address 99.99.99.99 255.255.255.255",
])
print(cfg_set_output)

# generic dispatcher (show vs config)
print(collector.execute_command("show version"))
print(collector.execute_command("logging buffered 8192", command_type="config"))

# save config
collector.backup_config()  # or collector.save_config()

collector.disconnect()
```
